```python
# Do best pixels stay best pixels?

# in regards to NW and resolution and is there any correlation?

from dataclasses import dataclass

@dataclass
class PixelData:
    temp: str
    pixel: tuple
    noise_width: float
    baseline: float
    resolution: float

pixels_data = []
for temp, noise_width_file in noise_widths.items():
    with open(noise_width_file) as file:
        nw_matrix = np.array(yaml.load(file, Loader=yaml.FullLoader))
    with open(baselines[temp]) as file:
        b_matrix = np.array(yaml.load(file, Loader=yaml.FullLoader))
    temp = float(temp[:-8])
    for row in range(16):
        for col in range(16):
            if beam_spots[temp][row, col]:
                res = full_data[temp].res_corr_map_data[row, col]
                if res > 0:
                    pixels_data.append(
                        PixelData(
                            temp=temp,
                            pixel=(row, col),
                            noise_width=nw_matrix[row, col],
                            baseline=b_matrix[row, col],
                            resolution=res
                        )
                    )
```

```python

def plot_heatmap_from_yaml(yaml_file, title=None):
    # Load the YAML file
    with open(yaml_file, 'r') as file:
        data = yaml.safe_load(file)
    
    # Create a heatmap using seaborn
    plt.figure(figsize=(6, 6))
    pixels_hist = Hist(
        #x axis = col
        hist.axis.Integer(0, 16, name="col", label="col", flow=False),
        #y axis = row
        hist.axis.Integer(0, 16, name="row", label="row", flow=False),
    )
    
    data = np.array(data)
    data[~beam_mask] = 0
    for (row, col), weight in np.ndenumerate(data):
        pixels_hist.fill(row=row, col=col, weight=weight)
        # x = col, y = row
        plt.text(col + 0.5, row + 0.5, f'{int(weight)}', ha='center', va='center', color='white' , fontsize="xx-small")
    hep.hist2dplot(pixels_hist)
    
    # Add title and labels if necessary
    plt.title(title)
    plt.xlabel('Column Index')
    plt.ylabel('Row Index')

# for name, path in noise_widths.items():
#     plot_heatmap_from_yaml(path, title=f'Noise Width: {name}')
```

```python
# find the leakage current values at different temperatures

"""
[2024-10-27T15:34:37]: [terter] bd [0] ch [1] par [VMon] val [2650.1]; 
[2024-10-27T15:34:37]: [terter] bd [0] ch [3] par [IMonH] val [26.245]; 
[2024-10-27T15:34:38]: [terter] bd [0] ch [1] par [VMon] val [2650]; 
[2024-10-27T15:34:38]: [terter] bd [0] ch [3] par [IMonH] val [26.23]; 
"""

t_twenty_degrees = ("Sat 26 Oct 2024 11:53:29 PM UTC", "Sun 27 Oct 2024 05:45:06 AM UTC")
t_twelve_degrees = ("Sun 27 Oct 2024 06:29:35 PM UTC", "Sun 27 Oct 2024 09:19:30 PM UTC")
t_four_degreees = ("Sun 27 Oct 2024 11:51:06 PM UTC", "Mon 28 Oct 2024 04:19:55 AM UTC")
t_minus_four_degrees = ("Wed 30 Oct 2024 12:54:44 AM UTC", "Wed 30 Oct 2024 03:09:49 AM UTC")
t_minus_twelve_degrees = ("Tue 29 Oct 2024 01:44:38 PM UTC", "Tue 29 Oct 2024 06:26:45 PM UTC")
t_minus_twenty_degrees = ("Thu 31 Oct 2024 03:01:35 AM UTC", "Thu 31 Oct 2024 05:30:47 AM UTC")


def extract_imonh_values(log_data, time_range):
    from datetime import datetime
    import re
    start_time = datetime.strptime(time_range[0], "%a %d %b %Y %I:%M:%S %p %Z")
    end_time = datetime.strptime(time_range[1], "%a %d %b %Y %I:%M:%S %p %Z")
    pattern = re.compile(r'\[(.*?)\]: \[terter\] bd \[0\] ch \[(3)\] par \[IMonH\] val \[(.*?)\];')
    
    imonh_values = []
    times = []
    for match in pattern.findall(log_data):
        try:
            log_time = datetime.strptime(match[0], "%Y-%m-%dT%H:%M:%S")
        except ValueError:
            log_time = datetime.now()

        if start_time <= log_time  and log_time <= end_time and match[1] == '3':
            times.append(log_time)
            imonh_values.append(float(match[-1]))
    return times, imonh_values

with open("/home/etl/CAENGECO2020.log", "r") as file:
    log_data = file.read()


times, current_values = extract_imonh_values(log_data, t_twelve_degrees)

times = np.array(times)
current_values = np.array(current_values)

dtimes = [t.seconds / 60 for t in times-times[0]]
print(dtimes)
plt.plot(dtimes, current_values)
plt.xlabel('Time (minutes)')
plt.ylabel('Leakage Current (uA)')
plt.title('12.5 C Leakage Current vs Time')

print("-------------")
print(len(current_values), times[-1] - times[0])
print(f"min: {min(current_values)} max: {max(current_values)} mean: {np.mean(current_values)}")
print("-------------")
```
