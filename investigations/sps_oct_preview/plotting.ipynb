```python
import hist
from hist import Hist
import matplotlib.pyplot as plt
import mplhep as hep
import json

with open("/home/etl/Test_Stand/tbanalysis/output_analysis_data/SPS_Oct_preview/start_10333_stop_10433_setup_110_offset_10.0_energy_120.0_power_i1_module_110_bias_245_file_from_SPS_Oct_LP2_20_cubicLM_unbinned/cal_mode.json") as f:
    cal_mode = json.load(f)
parquet = "/home/etl/Test_Stand/tbanalysis/output_analysis_data/SPS_Oct_preview/start_10333_stop_10433_setup_110_offset_10.0_energy_120.0_power_i1_module_110_bias_245_file_from_SPS_Oct_LP2_20_cubicLM_unbinned/fullCollection.parquet"
events = ak.from_parquet(parquet)

cal_array = ak.zeros_like(events.row)
for row in range(16):
    for col in range(16):
        pix_sel = ((events.row==row)&(events.col==col))
        cal_array = ak.where(pix_sel, cal_mode[row][col], cal_array) # think about cal_array = cal_mode[events.row, events.col]  
events["cal_mode"] = cal_array

clock_timestamps2 = events.Clock
time_bin = 3.125 / events.cal_mode
toa = 12.5 - time_bin * events.toa_code

good_clock = ak.flatten(clock_timestamps2[ak.num(toa)==1])
good_toa = ak.flatten(toa[ak.num(toa)==1])

hist_2d = Hist(
    hist.axis.Regular(100,-25,25, name='clock'),
    hist.axis.Regular(100, -25,25, name='toa'),
).fill(good_clock, good_toa)

hep.hist2dplot(hist_2d)
```
