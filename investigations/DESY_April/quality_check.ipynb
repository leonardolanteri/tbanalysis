```python
import numpy as np
import hist
from hist import Hist
import mplhep as hep
import matplotlib.pyplot as plt
plt.style.use(hep.style.CMS)
hits = np.zeros([16,16])

for tb in [BIAS_225_OFFSET_20_208]:
    tb_plot = tbplt.TBplot(tb)
    hits += tb_plot.hit_map_data

pixels_hist = Hist(
    #x axis = col
    hist.axis.Integer(0, 16, name="col", label="col", flow=False),
    #y axis = row
    hist.axis.Integer(0, 16, name="row", label="row", flow=False),
)
for (row, col), weight in np.ndenumerate(hits):
    pixels_hist.fill(row=row, col=col, weight=weight)
    # x = col, y = row
    plt.text(col + 0.5, row + 0.5, f'{int(weight)}', ha='center', va='center', color='white' , fontsize=10)
plt.title(f"209 Hits")
hep.hist2dplot(pixels_hist, cmax=45)
```
